### Categorical, Dummy Variables, One Hot Encoding

1. Using *pandas* **get_dummies**
2. Using *sklearn* **OneHotEncod**

**carprices.csv** \
This file has car sell prices for 3 different models. \
First plot data points on a scatter plot chart \
to see if linear regression model can be applied. \
If yes, then build a model that can answer \
following questions,

**1) Predict price of a mercedez benz that is 4 yr old with mileage 45000**

**2) Predict price of a BMW X5 that is 7 yr old with mileage 86000**

**3) Tell me the score (accuracy) of your model. (Hint: use LinearRegression().score())**

```python
import pandas as pd
from sklearn import linear_model
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('carprices.csv')
df
```

```python
df = df.rename(columns={'Car Model':'car_model','Sell Price($)':'sales_price','Mileage':'mileage','Age(yrs)':'age'})
df
```

```python
plt.plot(df.mileage,df.sales_price)
```

```python
dummies = pd.get_dummies(df.car_model)
```

```python
merge = pd.concat([df,dummies],axis='columns')
```

```python
final = merge.drop(['car_model','Audi A5'],axis='columns')
final.head(5)
```

```python
final = final.rename(columns={'BMW X5':'bmw','Mercedez Benz C class':'mbenz'})
final
```

```python
x = final.drop(['sales_price'],axis='columns')
x
```

```python
y = final.sales_price
```

```python
reg = linear_model.LinearRegression()
reg.fit(x,y)
```

```python
reg.predict([[45000,4,0,1]])
```

```python
reg.predict([[86000,7,1,0]])
```

```python
reg.score(x,y)
```
