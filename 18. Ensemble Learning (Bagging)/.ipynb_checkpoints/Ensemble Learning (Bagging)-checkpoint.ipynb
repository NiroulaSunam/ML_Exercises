```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
# Create some data
x = df.HeartDisease
y1 = df.Age
y2 = df.RestingBP
y3 = df.Cholesterol
y4 = df.FastingBS
y5 = df.MaxHR
y6 = df.Oldpeak

# Create a figure with two subplots
fig, axs = plt.subplots(3, 2, figsize=(10, 5))

# Plot the first subplot
axs[0, 0].plot(x, y1, 'g+')
axs[0, 0].set_title('Sine function')

# Plot the second subplot
axs[0, 1].plot(x, y2, 'b+')
axs[0, 1].set_title('Cosine function')

axs[1, 0].plot(x, y3, 'r+')
axs[1, 0].set_title('Sine function')

axs[1, 1].plot(x, y4, 'y+')
axs[1, 1].set_title('Sine function')

axs[2, 0].plot(x, y5, 'b+')
axs[2, 0].set_title('Sine function')

axs[2, 1].plot(x, y6, 'r+')
axs[2, 1].set_title('Sine function')

#Remove the last sublot (if there is not subplot in the last columns)
# fig.delaxes(axs[1, 1])

# Add a common y-axis label
fig.text(0.04, 0.5, 'Amplitude', ha='center', va='center', rotation='vertical')

# Show the plots
plt.show()
```
